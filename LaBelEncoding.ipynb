```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
df = pd.read_csv('hearti.csv' , header =None)
```

```python
df
```

```python
df.isnull().sum()
```

```python
encoder = LabelEncoder()
```

```python
dataset = df.iloc[1:].copy()
dataset.columns = df.iloc[0]
dataset = dataset.reset_index(drop=True)
```

```python
display(dataset.head())
display(dataset.info())
```

```python
categorical_cols = ['anaemia', 'diabetes', 'high_blood_pressure', 'sex', 'smoking', 'DEATH_EVENT']

for col in categorical_cols:
  dataset[col] = encoder.fit_transform(dataset[col])

display(dataset.head())
```

```python
encoded_sex = encoder.fit_transform(dataset['sex'])
display(encoded_sex)
```

```python
display(dataset['sex'].value_counts())
```

```python
dataset = dataset.drop(columns=['sex'])
display(dataset.head())
```
